# Photons produced by Scintillation

In this notebook, we will study the spectrum of the photons produced by scintillation after the passage of the muon through the scintillator.

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data file

file = "../build/output/step_output_Scintillator.txt"

photon_id, energy, wavelength, time = np.loadtxt(file, usecols=[0, 1, 2, 3], unpack=True)
```

```python
plt.figure()
plt.hist(energy, 51)
plt.xlabel("Energy (keV)")
plt.ylabel("Number of photons")
plt.show()
```

```python
plt.figure()
plt.scatter(time, wavelength)
```
